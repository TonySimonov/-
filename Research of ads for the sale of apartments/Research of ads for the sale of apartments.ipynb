# Исследование объявлений о продаже квартир
## Описание исследования
В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

### Цель исследования
Провести анализ рынка недвижимости, в частности квартир и выявить параметры влияющие на рыночную стоимость квартиры.

### Исходные данные
Датасет с информацией об объявлениях по продаже квартир `'real_estate_data.csv'`.

## Изучение данных из файла

```python
# Подключение библиотеки pandas
import pandas as pd
import matplotlib.pyplot as plt

# Отключение ошибок
pd.options.mode.chained_assignment = None  # default='warn'
pd.set_option('display.float_format', '{:,.2f}'.format)

# Чтение файла с данными
flats = pd.read_csv('/datasets/real_estate_data.csv', sep='\t')

# Вывод первых 10 строк датасета и общей информации
display(flats.head(10))
flats.info()
```

**Вывод**

Предоставленные данные состоят из 22 колонок и 23699 строк, среди данных встречаются пропуски.

## Предобработка данных

### Обработка пропусков

Для поиска пропусков воспользуемся методом `'isna()'` и подсчитаем пропуски при помощи метода `'sum()'`.

```python
(flats.isna().mean()*100).round(decimals=2)
```

В данных довольно много пропусков, если попытаться "почистить" данные, то можно лишиться почти всего датасета. Можно удалить те строки, где количество пропусков в колонках мало, например в колонках `'locality_name'` и `'floors_total'`. Суммарно данные из строк с пропусками в этих колонках составляют ~0.5% от всего датасета, что не повлияет на результат исследования.

```python
flats.dropna(subset=['locality_name', 'floors_total'], inplace=True)
# Проверка результата
(flats.isna().mean()*100).round(decimals=2)
```

От пропусков в столбцах  `'locality_name'` и `'floors_total'` избавились. Для остальных колонок можно попробовать заполнить пропуски данными, например, если не указано есть балкон или нет, то скорее всего балкона нет и значения можно заполнить нулями, аналогичную замену можно предложить для колонки `'is_apartment'`, если значения нет, то скорее всего это не аппартаменты.

```python
flats['balcony'].fillna(0, inplace=True)
flats['is_apartment'].fillna(False, inplace=True)
# Проверка результата
(flats.isna().mean()*100).round(decimals=2)
```

Если обратить внимание на колонку `'parks_nearest'` и `'ponds_nearest'`, то можно увидеть, что отсутствует большая часть данных в этих колонках. Скорее всего это связано с тем, что люди не знают даже примерное рассояние до ближайщего парка или пруда, заполнить пропуски какими-либо логичными значениями вряд ли получится, проще будет удалить эти колонки.

```python
flats.drop(columns=['parks_nearest','ponds_nearest'], inplace=True)
# Проверка результата
flats.info()
```

Теперь подробнее рассмотрим колонку `'ceiling_height'`. Диапазон значений не должен быть слишком большим, значит пропущенные значения можно заменить на медианой.

```python
# Убедимся в предположении
flats['ceiling_height'].describe()
```

```python
# Заменим пропуски медианой
flats['ceiling_height'].fillna(flats['ceiling_height'].median(), inplace=True)
# Проверка результата
(flats.isna().mean()*100).round(decimals=2)
```

Рассмотрим колонки `'living_area'` и `'kitchen_area'`. Жилая площадь и кухня, составляют некоторую часть от общей площади, можно связать эти параметры с количеством комнат и вместо пропусков вставить медианные значения в заисимости от количества комнат.

```python
# Замена значений жилой площади
flats['living_area'].fillna(flats.groupby('rooms')['living_area'].transform('median'), inplace=True)
# Замена значений для площади кухни
flats['kitchen_area'].fillna(flats.groupby('rooms')['kitchen_area'].transform('median'), inplace=True)
# Проверка
display(flats.head())
```

Остальные пустые значения лучше оставить как есть.

**Вывод**

В ходе обработки пропусков избавились от пустых значений в колонках  `'locality_name'` и `'floors_total'`, т.к. суммарное количество пропусков не влияет на датасет. В колонках `'balcony'` и `'is_apartment'` пропуски заменены на '0' и 'False' соответственно, сделано это исходя из следующей логики:
* если нет данных о балконах, то скорее всего их нет
* если нет данных о том аппартаменты это или нет, скорее всего данная квартира не относится к аппартаментам.

Пропуски в других колонках связаны с тем, что человек, выложивший объявление, скорее всего не знает об этих данных.

### Поиск дубликатов

Для поиска дубликатов воспользуемся методом `'duplicated()'`.

```python
flats.duplicated().sum()
```

Неявные дубликаты могут находится в колонке  `'locality_name'`, посмотрим какие на значения в этой колонке, и найдем количество уникальных значений.

```python
print(flats['locality_name'].unique())
len(flats['locality_name'].unique())
```

Есть различные написания слова "поселок" через "е" и через "ё", это приводит к появлению дубликатов. Воспользуемся методом replace().

```python
flats['locality_name'] = flats['locality_name'].str.replace('посёлок', 'поселок')
# Проверка
print(flats['locality_name'].unique())
len(flats['locality_name'].unique())
```

Значения со словами "посёлок" исчезли и количество уникальных значений стало меньше, значит от дубликатов избавились.

**Вывод**

Избавились от дубликатов в данных.

### Аномальные значения

Аномальные значения будут определены и учтены далее в исследовании при построении гистограмм. 

### Изменение типов данных

Переименуем колонку с дальностями до центра города в соответствии с "правилами хорошего тона".

```python
flats = flats.rename(columns = {'cityCenters_nearest' : 'city_center_nearest'})
```

Можно изменить тип данных в колонке `'first_day_exposition'`, т.к. это дата, а в таблице записан как строка. Учтем, что данные представлены в формате UTC, а время в Питере UTC + 3 часа. Округлим данные до 1 часа.

```python
flats['first_day_exposition'] = ((pd.to_datetime(flats['first_day_exposition'], format='%Y.%m.%dT%H:%M:%S')
                                 + pd.Timedelta(hours=3)).dt.round('1H'))
# Проверка результата
flats.info()
```

Тип данных колонки `'floors_total'` float, заменим его на int, т.к. количество этажей должно быть целым числом. 

```python
flats['floors_total'] = flats['floors_total'].astype('int')
# Проверка результата
flats.info()
```

**Вывод**

Тип данных в колонке `'first_day_exposition'` изменени на тип datetime64, в колонке `'floors_total'` на тип int.

## Расчёты и добавление результатов в таблицу

### Расчет цены за квадратный метр

В исходном датасете создадим колонку `'price_square'`, в которой будут хранится данные о цене за квадратный метр для каждой квартиры. 

```python
flats['price_square'] = flats['last_price'] / flats['total_area']
# Вывод первых 10 строк датасета и общей информации
display(flats.head())
flats.info()
```

### Определение даты публикации объявления

Отдельно введем 3 колонки: день недели, месяц, год продажи квартиры.

```python
# Колонка для дня недели
flats['weekday'] = flats['first_day_exposition'].dt.weekday
# Колонка для месяца
flats['year'] =  flats['first_day_exposition'].dt.strftime('%Y').astype('str').astype('int')
# Колонка для года
flats['month'] =  flats['first_day_exposition'].dt.strftime('%m').astype('str').astype('int')
# Проверка результата
display(flats.head())
```

### Определение этажа

Для описания этажа квартиры используется 3 варианта: первый, последний и другой. Напишем функцию, которая будет категоризировать данные по этажам.

```python
def group(row):
    floor = row['floor']
    last_floor = row['floors_total']
    
    if floor == 1:
        return 'первый'
    if floor == last_floor:
        return 'последний'
    return 'другой'

# Вызываем функцию и применяем ее к строке
flats['floor_group'] = flats.apply(group, axis=1)
# Проверка результата
display(flats.head())
```

### Соотношение жилой площади к общей

В датасете создадим колонку `'living_ratio'`, в котором будут содержаться данные, характеризующие отношение жилой площади к общей.

```python
flats['living_ratio'] = flats['living_area'] / flats['total_area']
# Проверка результата
display(flats.head())
```

### Соотношение кухни к общей площади

В датасете создадим колонку 'kitchen_ratio', в котором будут содержаться данные, характеризующие отношение кухни к общей площади.

```python
flats['kitchen_ratio'] = flats['kitchen_area'] / flats['total_area']
# Проверка результата
display(flats.head())
```

**Вывод**

На данном этапе дополнили исходный датасет следующими столбцами:
* price_square - цена за квадратный метр
* weekday - день недели, месяц и год публикации объявления
* floor_group - категория этажа
* living_ratio - отношение жилой площади к общей
* kitchen_ratio - отношение кухни к общей площади

Построим гистограммы для основных параметров: цена, площадь, число комнат, высота потолков. Для этого воспользуемся методом hist(), параметры bins и range которого подберем таким образом, чтобы детализировать распределение значений.

Основной пик приходится на квартиры с площадью 30-50 кв.м. В качестве больших можно принять квартиры с площадью свыше 200 кв.м., маленькими можно считать квартиры с площадью менее 25 кв. м. 

```python
# Гистограмма числа комнат
flats.plot(y='rooms', kind='hist', bins=20, title='Количество комнат', grid=True, legend=False)
plt.xlabel('Количество комнат')
plt.ylabel('Количество квартир')
```

Основной пик приходится на квартиры с 1-3 комнатами. В качестве больших можно выделить квартиры с более чем 6 комнатами, также следует учесть, что в данных есть квартиры с 0 комнат (возможно, при составлении объявления подразумевали 1 комнату).

```python
# Гистограмма высоты потолков
flats.plot(y='ceiling_height', kind='hist', bins=100, range=(0,30), title='Высота потолков', grid=True, legend=False)
plt.xlabel('Высота потолка (м)')
plt.ylabel('Количество квартир')
```

На этапе изучения основных параметров квартир обнаружены аномалии, а также выделены границы пригодных для дальнейшего исследования параметров.

### Оценка стоимости жилья в топ-10 населенных пунктах по количеству объявлений

Выделим 10 населенных пунктов с наибольшим числом объявлений, и рассчитаем среднюю цену за квадратный метр по каждому из них.

```python
# Ищем 10 населенных пунктов с наибольшим числом объявлений
top_10_index = good_sell['locality_name'].value_counts().head(10)

# Вырезаем из всех данных только те, чьи наименования воспадают с наименованиями из топ-10 по кол-ву объявлений
top_10 = good_sell.query('locality_name in @top_10_index.index')

# Создаем сводную таблицу, отображающую среднюю цену за квадратный метр по каждому населенновму пункту из топа
mean_price_top_10 = top_10.pivot_table(
    index='locality_name', values='price_square', aggfunc='mean').sort_values('price_square', ascending=False)
display(mean_price_top_10)
```

**Вывод**

На основании таблицы, можно сказать, что к населенным пунктам с самыми высокими ценами за кв.м. являются Санкт-Петербург и Пушкин, самая низкая цена за кв.м. в Выборге.

### Определение квартир, относящихся к центру города

Рассмотрим подробнее квартиры, продававшиеся в Санкт-Петербурге воспользуемся методом query(). Рассчитаем среднюю цену за километр.

```python
sp_flats = good_sell.query('locality_name == "Санкт-Петербург"')
# Округлим расстояния до центра Петергбурга до целых значений
sp_flats['rounded_dist'] = sp_flats['city_center_nearest'].round(decimals=0)

# Рассчитаем среднее значение цены на каждый километр
# введем список для хранения средних значений
price_per_km = []

# Запустим цикл по i от 1 до 31 (максимальное расстояние максимальное значение ~29,5 км)
# в цикле будем добавлять в список средние значения 
for i in range(1, 31):
    price_per_km.append(sp_flats.query('@i-1 <= rounded_dist/1000 < @i')['last_price_mil'].apply('mean'))

# Конвертация списка в series
price_per_km = pd.Series(price_per_km)

# Построение графика
price_per_km.plot(style='o-',grid=True)
plt.xlabel('Удаленность (км)')
plt.ylabel('Цена за км (млн)')
```

**Вывод**

Судя по графику после 5 км наблюдается падение цены. Центральная зона города - зона с радиусом 5 км от центра.
